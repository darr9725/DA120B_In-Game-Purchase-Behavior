```python
import matplotlib.pyplot as plt

players['age'].hist(bins=20)

plt.title('Age Distribution of Players')
plt.xlabel('Age')
plt.ylabel('Number of Players')

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

players['age'].dropna().hist(bins=20)

plt.title('Age Distribution of Players')
plt.xlabel('Age')
plt.ylabel('Number of Players')

plt.show()
```
