```python
import pandas as pd

players = pd.read_csv('/content/drive/MyDrive/datasets/players_cleaned_of_mobile_game_inapp_purchases.csv')
purchases = pd.read_csv('/content/drive/MyDrive/datasets/purchase_cleaned.csv')

print("Players loaded:", players.shape)
print("Purchases loaded:", purchases.shape)
```

```python
print(players['spending_segment'].value_counts())
```

```python
print(players['is_purchaser'].value_counts())
```

```python
payment_revenue = purchases.groupby('payment_method')['purchase_amount'].sum().sort_values(ascending=False)

print(payment_revenue)
```

```python
import matplotlib.pyplot as plt

players['spending_segment'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%'
)

plt.title('Player Spending Segments')
plt.ylabel('')

plt.show()
```

```python
players['is_purchaser'].value_counts().plot(
    kind='bar'
)

plt.title('Purchasers vs Non-Purchasers')
plt.xlabel('Purchaser Status')
plt.ylabel('Number of Players')

plt.show()
```

```python
payment_revenue.plot(
    kind='bar',
    figsize=(8,5)
)

plt.title('Revenue by Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Revenue (€)')

plt.show()
```
